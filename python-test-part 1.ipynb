```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from nsepy import get_history
from datetime import date
```

```python

sbin = get_history(symbol='SBIN',
                    start=date(2015,1,1), 
                    end=date(2016,1,1))
sbin.head()
```

```python
sbin[['Close']].plot()
```

```python
#plot
plt.figure(figsize=(52,4))
plt.plot(sbin['Close'], label='Prev lose')
```
